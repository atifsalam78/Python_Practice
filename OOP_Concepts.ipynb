#### Magic Method (Dunder Method)

* Are special kind of method and has it's own super power.

````python
class Student:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return """
        Student Name: {}
        Student Age:  {}
        """.format(self.name, self.age)


s1 = Student("Atif Salam", 45)
# print(s1) # without __str__ magic method output === > <__main__.Person object at 0x7fbfe4d8d570>
print(s1) # with __str__ magic method output === > Student Name: Atif Salam Student Age: 45
````

#### Instance Variable
* Instance variable is an object variable.
* Attribute created inside the constructor.
* The value of instance variable is different for every object, means value varies from object to object.
  * e.g: Student Name, Age, Gender etc
* Every object has its copy of the instance attribute.
* Objects do not share instance variables.
* This is a special variable whose value depends on the object.
* Using instance variable we can not implement counter.
* We can modify values by using the object reference.
    * Note: When you change the instance variables values of one object, the changes will not be reflected in the remaining objects because every object maintains a separate copy of the instance variable.

````python
class Student:
    
    def __init__(self, name, age, gender):
        self.student_name = name # Instance variables
        self.student_age = age # Instance variables
        self.student_gender = gender # Instance variables
````

#### Static or Class Variable

* Static variable is a class variable.
Syntax:
      ClassName.variable_Name
* e.g. University Name, Course Name etc.
* Created inside the class but outside all the methods.
* Class variables are ```shared``` by all instances of a class.
* Value is same for every object.
* We can use with `ClassName.ClassVariable`.
* we can also use with `self`, `self.ClassVariable`.
* We can call outside the class by instance of the class.
* We can modify the value of the class variable
    * Note: It is recommended to change the class variables value using the class name.
    * If you modify a class variable using an instance, it doesnt change the class variable itself. Instead, it creates an instance variable with the same name that shadows (or hides) the class variable for that instance.


````python
class Student:
    roll_no = 0 # This is static or class variable
    batch_id = "V"
    uni_name = "NED University of Engineering And Technology"
    
    def __init__(self, name, age, gender):
        self.student_name = name
        self.student_age = age
        self.student_gender = gender
        Student.roll_no += 1
        
    def display_info(self):
        print("""
        {}
        Batch          : {}
        Student Roll No: {}
        Student Name   : {}
        Studnet Age    : {}
        Student Gender : {}
        """.format(Student.uni_name, Student.batch_id, Student.roll_no,
                   self.student_name, self.student_age,
                  self.student_gender))
        print("Student's Roll No: {}, with self".format(self.roll_no))
        
s1 = Student("Arshah Shah", 45, "Male")
s1.display_info()
s2 = Student("Faisal Shah", 43, "Male")
s2.display_info()
s3 = Student("Faiza Atif", 40, "Female")
s3.display_info()
print("Student's Roll No: {}".format(Student.roll_no)) # Can be use outside the class


````

# Encapsulation
```
* The principle of hiding uncessary details from outside the world.
* Binding of methods and variables
* Using getter() and setter() method we give access to the outer world according to program's logic
* We can't use private variables/methods outside the class but can be use inside the class that's why we use getter/setter methods
  * __private_variable
  * __private_methods
  * In memory (_Class_name__private_variable)

```

#### Why nothing is truly private in Python?
```
* Because Python is a launguage made for 'Adults'.
* If any programmer declare his/her variables/methods private then it's a gentleman's agreement that other programer stay away from these.
* The concept is Python gives you option to hide from the outer world and nobody can guess what is the name of your private variable until and unless he/she check your class code.
```

````python
class Student:
    
    def __init__(self, name, age):
        self.__name = name
        self.age = age
        
    def get_name(self): # Getter Method
        return self.__name
        
    def set_name(self, new_name): #Setter Method
        self.__name = new_name
        return self.__name
    
    def display_detail(self):
        print("""
        Student Name: {}
        Student Age:  {}
        """.format(self.get_name(), self.age))    
    
    
s1 = Student("Ayesha Atif", 14)
s1.display_detail() # Display before changes in private variable using getter method
s1.get_name()
s1.set_name("Manahil Atif")
s1.display_detail() # Display after changes in private variable using getter/setter method
````

#### Static Methods (Utility Functions)
```
* The methods for which we access without creating an object of that method we can use by class name
Syntax:
      ClassName.static_method_name
* These are also called utility function
```

````python
class Student:
    
    counter = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Student.counter += 1
    
    @staticmethod
    def get_counter():
        return Student.counter
    
s1 = Student("Manahil Atif", 14)
print(Student.get_counter())

s2 = Student("Ayesha Atif", 7)
print(Student.get_counter())
````

# Inheritance
```
* It's not a software concept it comes from the nature.
* Child class can be use parent's class data.
* Inheritance to be use for code reusability.
* Child class object can't access parent's class private attirbutes and methods.
* Parent class can't access child class attributes and methods.
* If child class has it's own constructor than parent's class constructor would never be executed.
* If child class have not it's own constructor than parent's class constructor would be executed.
Syntax:
      ChildClassName(ParentClassName)
```
#### What's get inherited?
```
* Constructor
* Non-private Attritbutes
* Non-private methods
```
#### Method Overriding
```
* If there is inherittance and child class has it's own method than it's method would be called.
* If parent's and child class have same name method than child class method would be called.
```
#### Super Keyword
```
* Super is separate entity an in-built function not the part of class that's why `self` is not to be used.
* Super always be used inside the child class.
* By using super keyword we can access parent's class methods.
* Super can't use parent's class attributes(variables).
* Super can't work outside the class.
Syntax:
        super().method_name()
```
#### Types of inheritance
```
1.) Single Inheritance
2.) Multilevel Inheritance
3.) Hierarchical Inheritance
4.) Multiple Inheritance
5.) Hybrid Inheritance (Mix of all obove)
```

#### Single Inheritance
```
Parent
  |_ Child

Syntax:
        class Parent
        class Child(Parent)
```

````python
class Father:
    
    def __init__(self, hair_color, eyes_color, skin_tone):
        self.hair_color = hair_color
        self.eyes_color = eyes_color
        self.skin_tone = skin_tone
        
class Child(Father):
    
    def look_like(self):
        print(f"{self.hair_color}, {self.eyes_color}, {self.skin_tone}")

obj = Child("Black", "Brown", "Fair")
obj.look_like()
````

#### Multilevel Inheritance
```
Grand Father
  |_ Father
       |_ Child

Syntax:
        class GrandFather
        class Father(GrandFather)
        class Child(Father)
```

````python
class GrandFather:
    
    def __init__(self, hair_color, eyes_color, skin_tone):
        self.hair_color = hair_color
        self.eyes_color = eyes_color
        self.skin_tone = skin_tone
        
class Father(GrandFather):
    
    def look_like(self):
        print(f"{self.hair_color}, {self.eyes_color}, {self.skin_tone}")
        
class Child(Father):
    
    def look_like(self):
        print(f"{self.hair_color}, {self.eyes_color}, {self.skin_tone}")

obj1 = Father("Black", "Brown", "Fair")
obj1.look_like()
obj2 = Child("Black", "Brown", "Fair")
obj2.look_like()
````

#### Hierarchical Inheritance
```
Parent
  |_ Child_1
  |_ Child_2

Syntax:
        class Parent
        class childOne(parent)
        class childTwo(parent
```

````python
class Parent:
    
    def __init__(self):
        self.hair_color = "Black"
        self.skin_tone = "Fair"
        self.eyes_color = "Brown"
        
class Son(Parent):
    
    def look_like(self):
        print(f"{self.hair_color}, {self.eyes_color}, {self.skin_tone}")
        
class Daughter(Parent):
    
    def look_like(self):
        print(f"{self.hair_color}, {self.eyes_color}, {self.skin_tone}")
        
son = Son()
son.look_like()
daughter = Daughter()
daughter.look_like()
    
````

#### Multiple Inheritance (Diamond Problem)
```
* This type of inheritance is not exist in another programming languages especialy in JAVA because of ambugity which is know as `Diamond Problem`.
* But Python resolve this problem by using `Method Resolution Order`.
* If both parent classes having same name method than method of 1st order class(Father) would be executed.

Father   Mother
  |_Child _|

Syntax:
        class Father
        class Mother
        class Child(Father, Mother)   
```

````python
class Father:
    
    def __init__(self):
        self.hair_color = "Black"
        self.skin_tone = "Brown"
        self.eyes_color = "Black"
        
class Mother:
    
    def __init__(self):
        self.hair_color = "Brown"
        self.skin_tone = "Fair"
        self.eyes_color = "Brown"    
        
class Child(Mother, Father):
    
    def look_like(self):
        print(f"{self.hair_color}, {self.eyes_color}, {self.skin_tone}")
        
child = Child()
child.look_like()
````

```

* Language automatically figure out which method is need to be called on the basis of output.

```

```
Input: [1, 2, 3] + [4, 5, 6]
Output: [1, 2, 3, 4, 5, 6] ==> Merge two lists
```

# Abstraction
```
* Hiding the construction details
* Can apply constraints of child class
* Can't create object of `Abstract Class`.
* Child class object can't be created if it haven't method that mentioned in abstract class having decorator `@abstractmethod`.
* In abstract method there is no piece of code written in abstract method.
Syntax:
        class Parent(ABC):

          @abstractmethod
          def security(self)
            pass

        class Child(Parent):
          def security(self):
            Print("Security code would be written here")
```
#### Abstract Base Class (ABC)
```
* A class have atleast one `abstract method`.
* Inherited by ABC class
```
#### Concrete Method
```
* Normal method written inside the class
```
#### Abstract Method (@abstractmethod)
```
* Special type of method written inside the abstract class having decorator `@abstractmethod` and there is no piece of code inside in it.
```

````python
from abc import ABC, abstractmethod 
class Student(ABC):
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    @abstractmethod
    def fees():
        pass
    
    def display_info(self):
        print(f"Student Name: {self.name}, Student Age: {self.age}")

class Admission(Student):
    
    def fees(self):
        print("Fees Paid")
        
    def display_info(self):
        print(f"Student Name: {self.name}, Student Age: {self.age}")
        
s1 = Admission("Manahil", 14)
s1.display_info()
````
