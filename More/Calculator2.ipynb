```python
# Calculator Version No. 0.03

import getpass, math
count=0
space=' '*30
print('%sCalculator Vesrion 0.03'%space)
mp={1:'Basic',2:'Scientific',3:'Unit Conversion'}
mp1={'+':'Addition','-':'Subtraction','*':'Multiplication','/':'Division','^':'Square'}
mp2={'sqrt':'Square Root','sin':'Sin','cos':'Cos','tan':'Tan','log':'Log'}
while count<3:
    
    usr=input('Enter Your User ID: ')
    pin=int(getpass.getpass('Enter Your Pin: '))
    if usr=='atif' and pin==1003:
        print('Authenticated!')
        
        # Mode Selection
        
        sel=mp[int(input('What you want to Perform? '))]
        if sel=='Basic': 
            
            #Basic Calculator
            
            print('%s*********************'%space)
            print('%sYour Basic Calculator'%space)
            print('%s*********************'%space)
            num1=float(input('First Number: '))
            sel1=mp1[input('Enter Your Selection: ')]
            num2=float(input('Second Number: '))                       
            if sel1=='Addition':
                print('Your Answer is:',num1+num2)
            elif sel1=='Subtraction':
                print('Your Answer is:',num1-num2)
            elif sel1=='Multiplication':
                print('Your Answer is:',num1*num2)
            elif sel1=='Division':
                print('Your Answer is:',num1/num2)
            elif sel1=='Square':
                print('your Answer is: ',num1**num2)
            
        elif sel=='Scientific':
            
            #Scientific Calculator
            
            print('%s*********************'%space)
            print('%sYour Scientific Calculator'%space)
            print('%s*********************'%space)
            num1=float(input('First Scientific Number: '))
            sel2=mp2[input('Enter Your Scientific Selection: ')]
            #num2=float(input('Second Scientific Number: '))
            if sel2=='Square Root':
                print('Your Answer is:',math.sqrt(num1))
            elif sel2=='Sin':
                print('Your Answer is:',math.sin(num1))
            elif sel2=='Cos':
                print('Your Answer is:',math.cos(num1))
            elif sel2=='Tan':
                print('Your Answer is:',math.tan(num1))
            elif sel2=='Log':
                print('Your Answer is:',math.log(num1))
            
        elif sel=='Unit Conversion':
            
            
            #Unit Conversion
            
            print('%s*********************'%space)
            print('%sYour Unit Conversion'%space)
            print('%s*********************'%space)   
            
        break
    else:
        print('Go Back And Re-enter')
        count+=1
```
